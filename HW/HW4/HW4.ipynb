# Question 1

$\begin{align}
f(x) &= 0.05x^4+0.1x^3-0.75x^2-x+3\\
f'(x) &= 0.2x^3+0.3x^2-1.5x-1\\
f''(x)&= 0.6x^2+0.6x-1.5
\end{align}$



## (a)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize


# Define the function
def f(x):
    return 0.05 * x**4 + 0.1 * x**3 - 0.75 * x**2 - x + 3

# Generate x values for plotting
x_vals = np.linspace(-10, 10, 400)
y_vals = f(x_vals)

# Plot the function
plt.plot(x_vals, y_vals, label='f(x) = 0.05x^4 + 0.1x^3 - 0.75x^2 - x + 3')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()

# Find the global minimum using scipy.optimize.minimize
result = minimize(f, x0=0)

# Plot the global minimum
plt.scatter(result.x, f(result.x), color='red', label=f'Global Minimum: ({result.x[0]:.2f}, {f(result.x)[0]:.2f})')
plt.legend()

plt.ylim(-1, 5)
# Show the plot
plt.show()
```

## (b)

```python
from scipy.optimize import minimize_scalar, newton

def df(x):
    return 0.2 * x**3 + 0.3 * x**2 - 1.5 * x - 1

def ddf(x):
    return 0.6 * x**2 + 0.6 * x - 1.5

def newtons_method(x0, num_iterations):
    x_values = [x0]
    for _ in range(num_iterations):
        x0 = x0 - df(x0) / ddf(x0)
        x_values.append(x0)
    return x_values
```

```python
# Set initial point and number of iterations
x0 = 5
num_iterations = 1000

# Run Newton's method
iterations = newtons_method(x0, num_iterations)

# Plot the function
x_vals = np.linspace(-10, 10, 400)
y_vals = f(x_vals)
plt.plot(x_vals, y_vals, label='f(x) = 0.05x^4 + 0.1x^3 - 0.75x^2 - x + 3')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Newton\'s Method Iterations')
plt.grid(True)

# Plot the iterations on the graph
plt.scatter(iterations, [f(x) for x in iterations], color='red', label='Newton\'s Method Iterations')
plt.legend()

plt.ylim(-1, 5)
# Show the plot
plt.show()
```

Yes, it converges to global minimum.

```python
# Set initial point and number of iterations
x0 = -1
num_iterations = 1000

# Run Newton's method
iterations = newtons_method(x0, num_iterations)

# Plot the function
x_vals = np.linspace(-10, 10, 400)
y_vals = f(x_vals)
plt.plot(x_vals, y_vals, label='f(x) = 0.05x^4 + 0.1x^3 - 0.75x^2 - x + 3')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Newton\'s Method Iterations')
plt.grid(True)

# Plot the iterations on the graph
plt.scatter(iterations, [f(x) for x in iterations], color='red', label='Newton\'s Method Iterations')
plt.legend()

plt.ylim(-1, 5)
# Show the plot
plt.show()
```

No, it does not converge to the global minimum. Newton's method can encounter problems when $f'(x)=0$ (this is what happened here) or $f''(x)$ is undefined.

We can fix the learning rate by adding a small positive term.

```python
def newtons_method(x0, num_iterations, epsilon=1e-8, jump = 5):
    x_values = [x0]
    for _ in range(num_iterations):

        if abs(df(x0)) < epsilon:    
            if ddf(x0) < 0:
                x0 = x0 - ( df(x0) + jump ) / ddf(x0) 
        else:
            x0 = x0 -  df(x0) / ddf(x0)
            
        x_values.append(x0)
    return x_values


x0 = 0.1
num_iterations = 1000
# Run Newton's method
iterations = newtons_method(x0, num_iterations)

# Plot the function
x_vals = np.linspace(-10, 10, 400)
y_vals = f(x_vals)
plt.plot(x_vals, y_vals, label='f(x) = 0.05x^4 + 0.1x^3 - 0.75x^2 - x + 3')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Newton\'s Method Iterations')
plt.grid(True)

# Plot the iterations on the graph
plt.scatter(iterations, [f(x) for x in iterations], color='red', label='Newton\'s Method Iterations')
plt.legend()

plt.ylim(-1, 5)
# Show the plot
plt.show()
```
